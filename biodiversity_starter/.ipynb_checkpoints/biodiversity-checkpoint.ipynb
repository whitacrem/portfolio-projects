# Biodiversity in National Parks

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load and Preview Data

```python
species = pd.read_csv('species_info.csv')
print(species.head())
```

```python
observations = pd.read_csv('observations.csv')
print(observations.head())
```

## Endangered Species in Each Category

```python
# Printing unique values in conservation status list

print(species.conservation_status.unique())
```

```python
# Change nan to 'No Concern'

species.fillna('No Concern', inplace=True)
print(species.groupby("conservation_status").size())
```

```python
# Group species together in categories and then by conservation status, eliminating those that are of no concern

conservationStatus = species[species.conservation_status != "No Concern"].groupby(["conservation_status", "category"])["scientific_name"].count().unstack()
print(conservationStatus)
```

```python
# Plot conservation status by category type

ax = conservationStatus.plot(kind='bar')
ax.set_title("Distribution of Species' Conservation Status")
ax.set_xlabel("Conservation Status")
ax.set_ylabel("Number of Species")
```

```python
# The number of birds that are species of concern seems to greatly outnumber all the other species.  
# If we eliminate birds from the chart we can see the proportions of the other categories more clearly.

minusBirds = species[(species.conservation_status != "No Concern") & (species.category != "Bird")].groupby(["conservation_status", "category"])["scientific_name"].count().unstack()
ax2 = minusBirds.plot(kind='bar')
ax2.set_title("Distribution of Species' Conservation Status (Omitting Birds)")
ax2.set_xlabel("Conservation Status")
ax2.set_ylabel("Number of Species")
```

```python
# Is there a particular species that are more likely to be in one of the concerning categories 
# (engangered, in recovery, and species of concern)?

concerningSpecies = species[(species.conservation_status == 'Endangered') | (species.conservation_status == 'Species of Concern') | (species.conservation_status == 'Threatened')].groupby("category")["scientific_name"].count()
print(concerningSpecies)
```

```python
plt.pie(concerningSpecies, labels=species.category.unique(), autopct='%0.0f%%')
```

```python
# What species makes up the largest percentage of endangered species?

endangered = species[species.conservation_status == 'Endangered'].groupby("category")["scientific_name"].count()
print(endangered)
```

### Conclusion

The number of birds that are species of concern seems to greatly outnumber all the other species.  When birds are eliminated, vascular plants and mammals in the species of concern are largely outnumbering all other species counts.  Mammals make up most of the endangered species (7 out of 16 species). However, when looking at all the most concerning categories ("Endangered", "Threatened" and "Species of Concern"), birds remain the overwhelming majority of species.
