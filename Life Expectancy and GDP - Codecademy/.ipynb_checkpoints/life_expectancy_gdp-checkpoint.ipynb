# Life Expectancy and GDP Data, Visualized
### Codecademy Data Analyst Career Path Portfolio Project

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

#Loading and previewing the data
df = pd.read_csv('all_data.csv')

#Rename life expectancy column to make it easier to call
df.rename(columns={'Life expectancy at birth (years)': 'Life expectancy'}, inplace=True)

print(df.head())
```

### Exploratory Data Analysis

Has life expectancy increased over time in these six nations?

```python
sns.lineplot(data=df, x='Year', y='Life expectancy', hue='Country')
plt.title('Life Expectancy over Time')
plt.show()
```

Has the GDP increased over time in these six nations?

```python
sns.lineplot(data=df, x='Year', y='GDP', hue='Country')
plt.title('GDP over Time')
plt.show()
```

Is there a correlation between GDP and life expectancy?

```python
ax1 = plt.subplot(2, 3, 1)
sns.scatterplot(data=df[df['Country'] == 'Chile'], x='GDP', y='Life expectancy')
ax1.set_title('Chile')

ax2 = plt.subplot(2, 3, 2)
sns.scatterplot(data=df[df['Country'] == 'China'], x='GDP', y='Life expectancy')
ax2.set_title('China')

ax3 = plt.subplot(2, 3, 3)
sns.scatterplot(data=df[df['Country'] == 'Germany'], x='GDP', y='Life expectancy')
ax3.set_title('Germany')

ax4 = plt.subplot(2, 3, 4)
sns.scatterplot(data=df[df['Country'] == 'Mexico'], x='GDP', y='Life expectancy')
ax4.set_title('Mexico')

ax5 = plt.subplot(2, 3, 5)
sns.scatterplot(data=df[df['Country'] == 'United States of America'], x='GDP', y='Life expectancy')
ax5.set_title('USA')

ax6 = plt.subplot(2, 3, 6)
sns.scatterplot(data=df[df['Country'] == 'Zimbabwe'], x='GDP', y='Life expectancy')
ax6.set_title('Zimbabwe')

plt.tight_layout()
plt.subplots_adjust(wspace=0.8)

plt.show()
```

What is the average life expectancy in these nations?

```python
sns.boxplot(data=df, x='Country', y='Life expectancy')
plt.xticks(rotation=45)
plt.show()
```

```python
average_life_expectancy = df.groupby('Country')['Life expectancy'].mean().reset_index()
average_life_expectancy.rename(columns={'Life expectancy':'Average Life Expectancy'}, inplace=True)
print(average_life_expectancy)
```

What is the average life expectancy compared to the average GDP of each country?

```python
average_GDP = df.groupby('Country')['GDP'].mean().reset_index()

#Add this new data to the average life expectancy data frame so it can be graphed.
average_life_expectancy['Average GDP'] = average_GDP['GDP']
print(average_life_expectancy)
```

```python
sns.scatterplot(data=average_life_expectancy, x='Average GDP', y='Average Life Expectancy')
plt.show()
```

### Conclusion

As GDP increases in a nation, life expectancy at birth increases a long with it.  Zimbabwe has the lowest GDP and its life expectancy is the lowest by far.  However, surprisingly, the life expectancy and GDP data from the other nations in this dataset suggests that there is not seem to be a significant correlation between nations with higher GDPs having higher life expectancy.
